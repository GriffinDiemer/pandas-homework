```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Optional: give the displayed data cleaner formatting

```python
items = purchase_data["Item ID"].value_counts()
unique_items = items.count()

unique_items

total_revenue = purchase_data["Price"].sum()
total_purchases = purchase_data["Price"].count()
average_price = total_revenue/total_purchases

average_price

total_purchases

total_revenue

purchase_analysis = {"Number of Unique Items": [unique_items],
                    "Average Price":[average_price], "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]
                    }
total_purchase_analysis = pd.DataFrame(purchase_analysis)
total_purchase_analysis["Average Price"] = total_purchase_analysis["Average Price"].astype(float).map("${:,.2f}".format)
total_purchase_analysis["Total Revenue"] = total_purchase_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
total_purchase_analysis
```

```python
unique_sn = purchase_data.drop_duplicates(subset="SN", keep= 'first', inplace=False)
gender_count = pd.DataFrame(unique_sn["Gender"].value_counts())
gender_count = gender_count.rename(
    columns={"Gender": "Total Count"})
gender_perc = (gender_count/player_count)*100

gender_percent = gender_perc["Total Count"].astype(float).map("{:,.2f}%".format)
gen_percent = pd.DataFrame(gender_percent)

gender_count['Percentage of Players'] = gen_percent
gender_count
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
unique_sn = purchase_data.drop_duplicates(subset="SN", keep= 'first', inplace=False)
age_count = pd.DataFrame(unique_sn["Age"])
bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

age_count["Total Count"] = pd.cut(age_count["Age"], bins, labels=group_names)



age_count = pd.DataFrame(age_count["Total Count"].value_counts())
age_count = age_count.sort_index()
age_percent = (age_count/player_count)*100

age_percent = age_percent["Total Count"].astype(float).map("{:,.2f}%".format)


age_count["Percentage of Players"] = age_percent
age_count
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase = purchase_data.groupby("Age")
age_reduced = purchase_data.loc[:, ["SN","Age", "Price"]]


new_age_sort = pd.DataFrame(purchase_data["Age"])
bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

new_age_sort["Purchase Count"] = pd.cut(new_age_sort["Age"], bins, labels=group_names)

new_age_sort = pd.DataFrame(new_age_sort["Purchase Count"].value_counts())
new_age_sort = new_age_sort.sort_index()


age_reduced["Age Groups"] = pd.cut(age_reduced["Age"], bins, labels=group_names)
age_reduced = age_reduced.sort_index()

age_unique_sn = age_reduced.drop_duplicates(subset="SN", keep= 'first', inplace=False)
age_unique_sn = age_unique_sn.groupby("Age Groups")
age_reduced = age_reduced.groupby("Age Groups")


age_average_person = age_unique_sn["SN"].count()
age_total_price = age_reduced["Price"].sum()

age_total_purchase = age_reduced["Price"].count()

age_average_purchase_price = age_total_price/age_total_purchase

age_average_purchase = age_total_price/age_average_person

age_total_price = age_total_price.astype(float).map("${:,.2f}".format)
age_average_purchase_price = age_average_purchase_price.astype(float).map("${:,.2f}".format)
age_average_purchase = age_average_purchase.astype(float).map("${:,.2f}".format)

new_age_sort["Average Purchase Price"] = age_average_purchase_price
new_age_sort["Total Purchase Value"]= age_total_price

new_age_sort["Avg Total Purchase Per Person"] = age_average_purchase

new_age_sort
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_reduced = purchase_data.loc[:, ["SN", "Price"]]

sn_grouped = sn_reduced.groupby(["SN"])


sn_item_count = sn_grouped.count()
sn_item_count = sn_item_count.rename(
    columns={"Price": "Purchase Count"})

sn_total_purchase = sn_grouped["Price"].sum()
sn_avg_purchase = sn_total_purchase/sn_item_count["Purchase Count"]

sn_avg_purchase = sn_avg_purchase.astype(float).map("${:,.2f}".format)
sn_total_purchase = sn_total_purchase.astype(float).map("${:,.2f}".format)
sn_item_count["Average Purchase Price"] = sn_avg_purchase
sn_item_count["Total Purchase Value"] = sn_total_purchase

sn_item_count = sn_item_count.sort_values("Purchase Count", ascending=False)
sn_item_count.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_reduced = purchase_data.loc[:, ["Item ID","Item Name", "Price"]]

item_grouped = item_reduced.groupby(["Item ID","Item Name"])

item_number_count = item_grouped.count()
item_number_count = item_number_count.rename(
    columns={"Price": "Purchase Count"})

item_total_purchase = item_grouped["Price"].sum()
item_avg_purchase = item_total_purchase/item_number_count["Purchase Count"]

item_number_count["Item Price"] = item_avg_purchase
item_number_count["Total Purchase Value"] = item_total_purchase

item_number_count = item_number_count.sort_values("Purchase Count", ascending=False)
item_number_count["Item Price"] = item_number_count["Item Price"].astype(float).map("${:,.2f}".format)
item_number_count["Total Purchase Value"] = item_number_count["Total Purchase Value"].astype(float).map("${:,.2f}".format)

item_number_count.head()
```
